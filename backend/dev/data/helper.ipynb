```python
import pandas as pd
import ast
```

```python
df = pd.read_csv('./challenge_source_data.csv')
df['risk'] = df['wear']*df['weather']*df['vegetation'] # calculate the risk
```

```python
df = df.iloc[: , 1:]
df['id']= range(1, len(df) + 1) # give an id
```

```python
df['geometry'] = df['geometry'].apply(lambda x: ast.literal_eval(x)) # change from str to dict
```

```python
df['geometry'].apply(lambda x: x['coordinates'])
```

```python
df['geometry'][44]
```

```python
# swapping long and lat
for k in range(150):
    if k == 44 or k == 109:                # 44 and 109 has very short length
        for f in range(2): 
            print(k)
            front = df['geometry'][k]['coordinates'][f][0]
            print(front)
            df['geometry'][k]['coordinates'][f][0] = df['geometry'][k]['coordinates'][f][1]
            df['geometry'][k]['coordinates'][f][1] = front
            print(df['geometry'][k]['coordinates'][f][0], df['geometry'][k]['coordinates'][f][1])
    else:
        for i in range(len(df['geometry'][k]['coordinates'])):
            for j in range(len(df['geometry'][k]['coordinates'][i])):
                front = df['geometry'][k]['coordinates'][i][j][0]
                df['geometry'][k]['coordinates'][i][j][0] = df['geometry'][k]['coordinates'][i][j][1]
                df['geometry'][k]['coordinates'][i][j][1] = front
```
